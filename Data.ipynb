```python
#Using the Matplot and seaborn [With the seaborn we can plot the kde that we can not do with the matplot]  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

mean_age = df['Age'].mean()
mode_age = df['Age'].mode()[0]
max_age = df['Age'].max()
min_age = df['Age'].min()

plt.figure(figsize=(12,6))
sns.histplot(df['Age'],bins = 30, color = 'skyblue', alpha = 0.7, kde= True)

plt.axvline(mode_age, color='red', linestyle='dashed', linewidth=2, label=f'Mode: {mode_age}')
plt.axvline(mean_age, color='green', linestyle='dashed', linewidth=2, label=f'Mean: {mean_age:.2f}')
plt.axvline(min_age, color='purple', linestyle='dashed', linewidth=2, label=f'Min: {min_age}')
plt.axvline(max_age, color='orange', linestyle='dashed', linewidth=2, label=f'Max: {max_age}')

plt.xlabel('Age', fontsize = 12)
plt.ylabel('Frequency ', fontsize =12)
plt.title('Age Distribution')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


mode_age = df['Age'].mode()[0]   
mean_age = df['Age'].mean()    
min_age = df['Age'].min()        
max_age = df['Age'].max()        


plt.figure(figsize=(12, 6))
plt.hist(df['Age'],edgecolor = 'Black', color='skyblue',alpha= 0.7,bins =30)

plt.axvline(mode_age, color='red', linestyle='dashed', linewidth=2, label=f'Mode: {mode_age}')
plt.axvline(mean_age, color='green', linestyle='dashed', linewidth=2, label=f'Mean: {mean_age:.2f}')
plt.axvline(min_age, color='purple', linestyle='dashed', linewidth=2, label=f'Min: {min_age}')
plt.axvline(max_age, color='orange', linestyle='dashed', linewidth=2, label=f'Max: {max_age}')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Age Distribution with Key Statistics', fontsize=14)
plt.legend()  

plt.show()




```

```python
# Crosstab between Blood Type and Gender (example)
crosstab_result = pd.crosstab(df['Blood Type'], df['Gender'])
print(crosstab_result)

# Heatmap to visualize the crosstab result
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab_result, annot=True, cmap='Blues', fmt='d')
plt.title('Blood Type vs Gender')
plt.show()
```

```python
# Example assuming there is a 'Date' column
df['Year'] = pd.to_datetime(df['Date of Admission']).dt.year
blood_by_year = df.groupby(['Year', 'Blood Type']).size().unstack()

blood_by_year.plot(kind='line', figsize=(12, 6))
plt.title('Blood Type Distribution Over Time')
plt.ylabel('Frequency')
plt.show()
```
