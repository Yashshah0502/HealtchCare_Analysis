```python
# Average age
avg_age = df.groupby('Medical Condition')['Age'].mean().reset_index()
avg_age.columns = ['Medical Condition', 'Average Age']

# Mode age: Custom function to get the first mode if multiple exist
mode_age = (
    df.groupby('Medical Condition')['Age']
    .agg(lambda x: x.mode().iloc[0])  # gets the first mode if multiple
    .reset_index()
)
mode_age.columns = ['Medical Condition', 'Mode Age']

# Combine both into one table (optional)
age_stats = avg_age.merge(mode_age, on='Medical Condition')

print(age_stats)
```

```python
# Average Age by Medical Condition and Gender
avg_age = (
    df.groupby(['Medical Condition', 'Gender'])['Age']
    .mean()
    .reset_index()
)
avg_age.columns = ['Medical Condition', 'Gender', 'Average Age']

# Mode Age by Medical Condition and Gender
mode_age = (
    df.groupby(['Medical Condition', 'Gender'])['Age']
    .agg(lambda x: x.mode().iloc[0])  # gets the first mode if multiple
    .reset_index()
)
mode_age.columns = ['Medical Condition', 'Gender', 'Mode Age']

# Merge both
age_stats = avg_age.merge(mode_age, on=['Medical Condition', 'Gender'])

print(age_stats)
```

```python
# Group by 'Medical Condition' and calculate the min, max, and mean of the 'Billing Amount'
medical_condition_stats = df.groupby('Medical Condition')['Billing Amount'].agg(['min', 'max', 'mean']).reset_index()

# Rename the columns for clarity
medical_condition_stats.columns = ['Medical Condition', 'Min Billing Amount', 'Max Billing Amount', 'Average Billing Amount']

# Print the result
print(medical_condition_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
medical_condition_stats = pd.DataFrame({
    'Medical Condition': ['Asthma', 'Cancer', 'Diabetes', 'Hypertension', 'Obesity'],
    'Min Billing Amount': [500, 1000, 600, 300, 400],
    'Max Billing Amount': [2500, 8000, 3500, 2000, 4000],
    'Average Billing Amount': [1450, 4500, 2100, 1200, 2300]
})

# Sort by Average Billing Amount for better visual comparison
order = medical_condition_stats.sort_values('Average Billing Amount')['Medical Condition']

# Reshape for plotting
df_melted = pd.melt(
    medical_condition_stats,
    id_vars='Medical Condition',
    value_vars=['Min Billing Amount', 'Max Billing Amount', 'Average Billing Amount'],
    var_name='Statistic',
    value_name='Billing Amount'
)

# Set plot style
sns.set(style="whitegrid", palette="Set2")
plt.figure(figsize=(12, 7))

# Barplot
ax = sns.barplot(
    x='Medical Condition',
    y='Billing Amount',
    hue='Statistic',
    data=df_melted,
    order=order
)

# Value labels
for p in ax.patches:
    value = p.get_height()
    ax.annotate(f'{value:,.0f}', 
                (p.get_x() + p.get_width() / 2, value),
                ha='center', va='bottom', fontsize=9)

# Labels and layout
plt.title('Billing Amount Statistics by Medical Condition', fontsize=16)
plt.xlabel('Medical Condition')
plt.ylabel('Billing Amount')
plt.xticks(rotation=45)
plt.legend(title='Statistic')
plt.tight_layout()
plt.show()
```
